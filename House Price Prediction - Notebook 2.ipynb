```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

from sklearn.preprocessing import MinMaxScaler

pd.pandas.set_option('display.max_columns',None)
```

```python
df_train = pd.read_csv("C:/Users/Priyaranjan/Desktop/Kaggle datasets/Advanced House Price Prediction/train.csv")
df_train.head()
```

### 2.1. Missing values Imputation

#### 2.11. Categorical features:
- creating a new label 'Missing' to replace the null values in categorical features.

```python
features_with_na = [features for features in df_train.columns if df_train[features].isnull().sum() > 1 and df_train[features].dtypes == 'O']

for features in features_with_na:
    print(features, np.round(df_train[features].isnull().mean(),4), '% missing values')
```

```python
def impute_cat_features(df_train,features_with_na):
    data = df_train.copy()
    data[features_with_na] = data[features_with_na].fillna('Missing')
    return data

df_train = impute_cat_features(df_train,features_with_na)
df_train[features_with_na].isnull().sum()
```

#### 2.12. Numerical features:
- Create a new feature & replace the null values in numerical features with their median values as we found outliers in our explarotary phase.

```python
numericals_with_na = [feature for feature in df_train.columns if df_train[feature].isnull().sum() > 1 and df_train[feature].dtypes != 'O']
for features in numericals_with_na:
    print(features, np.round(df_train[features].isnull().mean(),4), '% missing values')
```

```python
for feature in numericals_with_na:
    median = df_train[feature].median()
    df_train[feature+'nan'] = np.where(df_train[feature].isnull(),1,0)
    df_train[feature].fillna(median,inplace=True)

df_train[numericals_with_na].isnull().sum()
```

```python
df_train.head(100)
```

### 2.2. Handling Temporal variables (date-time variables):
- Convert the year columns data points into number of years representing relationship of the feature with 'YrSold'.

```python
for feature in ['YearBuilt', 'YearRemodAdd', 'GarageYrBlt']:
    df_train[feature] = df_train['YrSold']-df_train[feature]

df_train.head()
```

### 2.3. Standardizing the numerical features with outliers
From notebook 1, section 1.4. ; we know that we have outliers in columns: 'LotFrontageArea', 'LotArea', '1stFlrSF', 'GrLivArea', 'SalePrice'.

Hence, log normalizing the values in those columns to handle outliers.

```python
num_features = ['LotFrontage', 'LotArea', '1stFlrSF', 'GrLivArea', 'SalePrice']
for feature in num_features:
    df_train[feature] = np.log(df_train[feature])

df_train.head()
```

### 2.4. Categorical variables

#### 2.41. Handling rare labels
- We will be replacing the categories in categorical variables which make less than 1% of the total observations into a common category 'Rare_var'.

```python
categorical_features = [feature for feature in df_train.columns if df_train[feature].dtypes == 'O']
categorical_features
```

```python
for feature in categorical_features:
    temp = df_train.groupby(feature)['SalePrice'].count()/len(df_train)
    df_temp = temp[temp>0.01].index
    df_train[feature] = np.where(df_train[feature].isin(df_temp), df_train[feature], 'Rare_var')
    
df_train.head(50)
```

#### 2.42. Converting categorical into numerical labels.
- As, we would be feeding the final data into models, we would need numeric data to represent the categories.
    - for example: Class A assigned as Class 1, Class B assigned as Class 2.
- This assignment of numeric label is based on the sorted mean values of 'SalePrice' of each category in each categorical column, starting from 0.
    - ex: if Class a, b, c, d have mean 2.3, 4.5, 6.8, 2.1 of target variable respectively.
        - assignment would be based on sorted mean values.
        - class d : class 0, class a : class 1, class b : class 2, class c : class 3

```python
for feature in categorical_features:
    labels_ordered=df_train.groupby([feature])['SalePrice'].mean().sort_values().index
    labels_ordered={k:i for i,k in enumerate(labels_ordered,0)}
    df_train[feature]=df_train[feature].map(labels_ordered)

df_train.head(50)
```

```python
for cat in df_train['MSZoning'].unique():
    mean = df_train[df_train['MSZoning'] == cat]['SalePrice'].mean()
    print('Mean of class {} is {}'.format(cat,mean))
```

### 2.5. Feature Scaling:
- Now that we have all our data in numeric representation, we need to scale all the features as the features here might be of different ranges.
    - Scaling would be done on all features except 'Id' & 'SalePrice'( ***the target/dependent variable).***

```python
# All features to be scaled except the 'Id' & 'SalePrice'
scale_features = [feature for feature in df_train.columns if feature not in ['Id', 'SalePrice']]

# Initializing the object
scaler = MinMaxScaler()
scaler.fit(df_train[scale_features])
```

```python
# Transforming all the features except id & target variable & concatenating these 2 columns to the transformed result.
df_train = pd.concat([df_train[['Id', 'SalePrice']].reset_index(drop=True),
                    pd.DataFrame(scaler.transform(df_train[scale_features]),columns=scale_features)],
                    axis = 1)
df_train.head(50)
```

With the above step, we now have our final data to be sent for training on ML model. This data will be sent to staging area for model building action steps.

```python
# Downloading the final model data to be consumed in staging area
df_train.to_csv('model_data_train.csv',index=False)
```
